### DSP 2

Leakage se refiere a la perdida de informacion al cambiar los datos de analogos a digitales, esto ocurre debido a que las señales no coinciden debido a la frecuencia. Sin embargo esta perdida de informacion se puede contrarrestar usando funciones ventana. 

Las funciones ventana se caracterisan por su uso en el analisis de señales, estas representan una funcion matematica que mitiga o quita la perdida de esta informacion. 

```python
import numpy as np
from numpy import sin, cos, pi
import matplotlib.pyplot as plt

from matplotlib.pyplot import subplots
from scipy.fftpack import fft, ifft, fftshift
```

```python
fs = 64
ts = 1/fs
f1 = 3
f2 = 4
f3 = 5
f4 = 6

N = np.arange(fs)
x1 = sin(2*pi*f1*N*ts)
x2 = sin(2*pi*f2*N*ts)
x3 = sin(2*pi*f3*N*ts)
x4 = sin(2*pi*f4*N*ts)
```

```python
plots = lambda: plt.subplots(figsize = (20 , 10))
f, ax = plots()
ax.plot(N, x1, 'b*')
ax.plot(N, x2, 'g*')
ax.plot(N, x3, 'y*')
ax.plot(N, x4, 'r*')

plt.show()
```
