#### Bubble sort

```python
def bubble_sort(arr):
    # Flag to indicate the missing subarray has not been sorted
    ordered = False
    print(arr)
    while not ordered:
        ordered = True
        for i in range(1, len(arr)):
            
            if arr[i-1] > arr[i]:
                arr[i-1], arr[i] = arr[i], arr[i-1]
                ordered = False
                print(arr)
    print(arr)
```

```python
a = b.copy()
```

```python
bubble_sort(a)
```

#### Selection sort

```python
def selection_sort(arr):
    sorted_idx = 0
    print(arr)
    while sorted_idx < len(arr):
        for i in range(sorted_idx, len(arr)):
            if arr[sorted_idx] > arr[i]:
                arr[sorted_idx], arr[i] = arr[i], arr[sorted_idx]
                print(arr)
        sorted_idx += 1
    print(arr)
    
    
```

```python
a = [random.randint(0,100) for _ in range(20)]
print(a)
```

```python
selection_sort(a)
```

#### Insertion sort

```python
def insertion_sort(arr):
    print(arr)
    for j in range(1, len(arr)):
        #pick a value of the unsorted subarray to start comparing to sorted left subarray
        pivot = arr[j]
        # get index of previous value to pivot
        i = j-1
        # repeat until you get to first element of sorted array
        while (i > -1) and (arr[i]>pivot):
            # if pivot is smaller than previous in ordered array, put that value in position of pivot
            print(arr)
            arr[i+1] = arr[i]
            # move index left 
            i -= 1
            
        # assign pivot value to correct index
        arr[i+1] = pivot
        print(f'sorted so far {arr}')
    return arr
```

```python
insertion_sort(a)
```

#### Merge sort

```python
a = [random.randint(0,100) for _ in range(20)]
print(a)
```

```python
 def merge(list1, list2):
    result = []
    i, j = 0, 0 #indices, i for list1 and j for list2
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            result.append(list1[i])
            i+=1
        else:
            result.append(list2[i])
            j+=1
    #once some list is empty
    if i == len(list1):
        # append the rest of list 2 in result
        for k in range(j, len(list2)):
            result.append(list2[k])
    if j == len(list2):
        # append the rest of list2
        for k in range(i, len(list1)):
            result.append(list1[k])
    return result
```

```python
def merge_sort(l):
    if len(l) < 2:
        return l
    else:
        middle =  len(l)//2
        first_half = merge_sort(l[:middle])
        second_half = merge_sort(l[middle:])
        
        return merge(first_half, second_half)
```

#### Quicksort

```python
def quicksort(l):
    if len(l) < 2:
        return l
    else:
        #get pivot
        pivot = l[0]
        #get lower
        smaller_half = [i for i in l[1:] if i <= pivot]
        #get higher
        larger_half = [j for j in l[1:] if j > pivot]
        return quicksort(smaller_half) + [pivot] + quicksort(larger_half)
        
```
